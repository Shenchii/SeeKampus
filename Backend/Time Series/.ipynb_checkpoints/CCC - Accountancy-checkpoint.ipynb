```python
import pandas as pd
import prophet as Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
df = pd.read_csv('CSV/board exam results - CCC-Accountancy.csv')
df.head()
```

```python
df.dtypes
```

```python
df['Year'] = df['Time Date'].apply(lambda x: str(x)[-4:])
df['Month'] = df['Time Date'].apply(lambda x: str(x)[-6:-4])
df['Day'] = df['Time Date'].apply(lambda x: str(x)[:-6])
df['ds'] = pd.DatetimeIndex(df['Year']+'-'+df['Month']+'-'+df['Day'])
```

```python
df.dtypes
```

```python
df.drop(['Time Date', 'School', 'Course', 'Year', 'Month', 'Day'], axis=1, inplace=True)
df.columns = ['y', 'ds']
```

```python
df
```

```python
m = Prophet.Prophet(interval_width=0.95, yearly_seasonality=True)
model = m.fit(df)
```

```python
future = m.make_future_dataframe(periods=730,freq= "d")
forecast = m.predict(future)
forecast.head()
```

```python
forecast[['ds', 'yhat']]
```

```python
plot1 = m.plot(forecast)
```
